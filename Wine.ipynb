```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar a Base de Dados "Wine" direto da biblioteca
dados_wine = load_wine()

# Criar o DataFrame (Tabela)
df = pd.DataFrame(data=dados_wine.data, columns=dados_wine.feature_names)
df['target'] = dados_wine.target

print("--- Dados Carregados com Sucesso! ---")
print(f"Total de linhas: {len(df)}")
print(f"Total de colunas: {len(df.columns)}\n")


# Divisão entre Features (X) e Target (y)
X = df.drop('target', axis=1)  # Dados químicos
y = df['target']               # Tipo do vinho (0, 1, 2)

# Seleção de Características (Top 10)
print("Selecionando as 10 melhores características...")
selector = SelectKBest(score_func=f_classif, k=10)
X_selected = selector.fit_transform(X, y)

# Mostrar quais colunas ficaram
cols = selector.get_support(indices=True)
features_escolhidas = X.iloc[:,cols].columns.tolist()
print(f"Características Escolhidas: {features_escolhidas}\n")

# Divisão em Treino (70%) e Teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Padronização (StandardScaler)
# Importante: Fit no treino, Transform no treino e teste
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo A: SVM
print("Treinando SVM...")
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train_scaled, y_train)
y_pred_svm = svm_model.predict(X_test_scaled)

# Modelo B: Árvore de Decisão
print("Treinando Árvore de Decisão...")
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

# --- ETAPA 4: EXIBIR RESULTADOS ---

# Resultados SVM
print("\n" + "="*40)
print("RESULTADOS DO SVM")
print("="*40)
print("Acurácia:", accuracy_score(y_test, y_pred_svm))
print("\nRelatório Detalhado:")
print(classification_report(y_test, y_pred_svm))

# Gráfico SVM
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred_svm), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusão - SVM')
plt.ylabel('Verdadeiro')
plt.xlabel('Predito')
plt.show()

# Resultados Árvore
print("\n" + "="*40)
print("RESULTADOS DA ÁRVORE DE DECISÃO")
print("="*40)
print("Acurácia:", accuracy_score(y_test, y_pred_dt))
print("\nRelatório Detalhado:")
print(classification_report(y_test, y_pred_dt))

# Gráfico Árvore
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred_dt), annot=True, fmt='d', cmap='Greens')
plt.title('Matriz de Confusão - Árvore de Decisão')
plt.ylabel('Verdadeiro')
plt.xlabel('Predito')
plt.show()
```
